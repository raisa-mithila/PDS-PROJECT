```python
from sklearn.datasets import fetch_california_housing
data = fetch_california_housing()
df = pd.DataFrame(data.data, columns=data.feature_names)
df['target'] = data.target
```

```python
print(df.head())
print(df.describe())
sns.heatmap(df.corr(), annot=True, cmap="coolwarm")
plt.show()
```

```python
X = df.drop("target", axis=1)
y = df["target"]
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
model = LinearRegression()
model.fit(X_scaled, y)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

```python
y_pred = model.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
print(f"Mean Squared Error: {mse}")
print(f"R2 Score: {r2}")
```

```python
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
from sklearn.metrics import mean_squared_log_error
import numpy as np

y_pred = model.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)

# Ensure y_test and y_pred are non-negative for MSLE
y_test_msle = np.where(y_test < 0, 0, y_test)
y_pred_msle = np.where(y_pred < 0, 0, y_pred)

msle = mean_squared_log_error(y_test_msle, y_pred_msle)

print(f"Mean Squared Error: {mse}")
print(f"R2 Score: {r2}")
print(f"Mean Absolute Error: {mae}")
print(f"Mean Squared Log Error: {msle}")
```

```python
from sklearn.model_selection import cross_val_score
scores = cross_val_score(model, X_scaled, y, cv=5, scoring="r2")
print("Cross-Validation R2 Scores:", scores)
print("Mean Cross-Validation R2 Score:", np.mean(scores))
```
